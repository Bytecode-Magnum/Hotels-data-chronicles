```python
def scroll_windows():
    driver.execute_script('window.scrollTo(0,document.body.scrollHeight);')
def scroll():
    while True:
        old_height=driver.execute_script('return document.body.scrollHeight')
        scroll_windows()
        time.sleep(2)
        new_height=driver.execute_script('return document.body.scrollHeight');
        if new_height==old_height:
            break
```

#### Review Scraping

```python
def scrape_reviews():
    delimiter="*"
    soup=BeautifulSoup(driver.page_source,'html.parser')
    try:
        container=soup.find('div',class_='Layouts__Column-sc-1yzlivq-1 RnRstyles__UserReviewOuterWrapperDiv-sc-1vfl02s-9 fzWoxL cLkcPE')
        review_divs=container.find_all('div',recursive=False)
        for idx,div in enumerate(review_divs):
            try:
                reviews=(div.find('div',class_='Layouts__Column-sc-1yzlivq-1 UserReviewstyles__UserReviewWrapperDiv-sc-uxnwyv-3 fzWoxL gAmxQD').text.strip())+delimiter
                _data[f'{name}']['reviews'].append(reviews)
            except:
                pass
    except:
        pass
```

```python
def start_scraping_reviews():
    try:
        
        scrape_reviews()
        driver.find_element(By.XPATH,'//span[@class="Paginationstyles__RightDoubleSkipStyles-sc-1xuxk5q-5 ldMdQi"]').click()
        
        pagination_ul=driver.find_element(By.XPATH,"//ul[@class='Paginationstyles__PaginationUl-sc-1xuxk5q-9 hLzZXm']")
        last_li=pagination_ul.find_elements(By.TAG_NAME,'li')[-1]
        last_li=last_li.text
        last_li=int(last_li)
        scrape_reviews()
        
        for i in tqdm(range(last_li-1,1,-1),desc=f'Extracting reviews:-->'):
            left_btn=driver.find_element(By.XPATH,"//span[@class='Paginationstyles__LeftSkipStyles-sc-1xuxk5q-3 hAhRzV']")
            left_btn.click()
            time.sleep(2)
            scrape_reviews()
    except:
        pass
```

#### Aminities Scraping

```python
def start_Scraping_aminities():
    _data[f'{name}']['aminities']={}
    aminities_container=soup.find('div',class_='Amenitiesstyles__AmenitiesOuterWrapper-sc-e2msby-2 gyihyx')
    container=aminities_container.find('div',class_='Amenitiesstyles__CategorizedAmenitiesTableContainer-sc-e2msby-11 gkaJhd')
    _divs=container.find_all('div',recursive=False)
    for div in _divs:
        text=(div.find('h3').text)
        _data[f'{name}']['aminities'][f'{text}']=[]
        _ul=div.find('ul')
        _lis=_ul.find_all('li')
        for li in _lis:
            _data[f'{name}']['aminities'][f'{text}'].append(li.text)
```

#### Policies

```python
def start_scraping_policies():
      try:
            driver.find_element(By.XPATH, "//a[@class='dwebCommonstyles__PrimaryLink-sc-112ty3f-11 HouseRulesstyles__MorePolicyLink-sc-5t10ue-3 JGjqD eCxNdJ']").click()
            time.sleep(3)
            soup = BeautifulSoup(driver.page_source, 'html.parser')
            policies_div = soup.find('div', class_='HotelsModal__ModalGuts-sc-rgix69-2 fQQVoF')
            policies_lis = policies_div.find_all('li', class_='PolicyModalstyles__HouseRulesList-sc-825j2-2 kLqGdQ')
            for each in policies_lis:
                head = each.find('p', class_='PolicyModalstyles__PolicyListHeading-sc-825j2-4 IsRjr').text
                _data[f'{name}']['policies'][f'{head}'] = []
                ul = each.find('ul', class_='PolicyModalstyles__FeatureListUL-sc-825j2-6 gboSFx')
                lis = ul.find_all('li')
                for li in lis:
                    txt = li.text
                    _data[f'{name}']['policies'][f'{head}'].append(txt)
      except:
            pass
            
```

#### Room types

#### fooding section

```python
def start_scraping_fooding():
    try:
        _data[f'{name}']['fooding-rules']=[]
        food_dining=soup.find('section',class_='SingleRestaurentUIstyles__SingleRestaurentUI-sc-lkaikv-0 YiZTi')
        food_divs=food_dining.find('div',class_='DiningDetailsUIstyles__DiningWrapper-sc-qu13qq-0 cpUFjZ')
        row_divs=food_divs.find_all('div',recursive=False)
        for each_row in row_divs:
            for each in each_row.find_all('div',recursive=False):
                    _data[f'{name}']['fooding-rules'].append(each.text)
    except:
        pass
```

#### Aggregating everything

```python
hotel_data=[]
_data={}
for idx,value in tqdm(df.iterrows(),desc='Data Extraction in process...........'):
    driver.refresh()
    link=value['Link']
    name=value['Name']
    driver.get(link)
    time.sleep(3)
    #: reviews
    _data[f'{name}']={}
    _data[f'{name}']['reviews']=[]
    start_scraping_reviews()
    #: aminities
    _data[f'{name}']['aminities']={}
    start_Scraping_aminities()
    #: room types
    _data[f'{name}']['roomtypes']={}
    start_scraping_room_types()
    #: fooding
    _data[f'{name}']['fooding-rules']=[]
    start_scraping_fooding()
    #: policies
    _data[f'{name}']['policies']={}
    start_scraping_policies()

    
```
